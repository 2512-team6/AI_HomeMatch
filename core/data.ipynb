```python
import requests
import json
```

```python
BASE_URL = "http://www.law.go.kr/DRF/lawSearch.do"
```

```python
OC = "dgjjanggu"
LAW_NAME = "주택임대차보호법"
```

```python
params = {
    "OC":"dgjjanggu",
    "target":"eflaw",
    "type":"JSON",
    "query": LAW_NAME,
    "nw":3,
    "display":1,
    "page":1,
}

res = requests.get(BASE_URL, params=params, timeout=30)
res.raise_for_status

data = res.json()
```

```python
a = json.dumps(data, indent=2, ensure_ascii=False)
```

```python
print(a)
```

```python
url = "http://www.law.go.kr/DRF/lawSearch.do?target=prec"
QUERY = "임차인"
params = {
    "OC": OC,
    "target": "prec",
    "type": "JSON",
    "search": 2,        # 1=판례명, 2=본문검색 (원하면 1로 바꿔도 됨)
    "query": QUERY,
    "display": 20,      # 1~100
    "page": 1,
    "sort": "ddes",     # 선고일자 내림차순(기본이긴 한데 명시)
    # "org": "400201",  # 대법원만 보고 싶으면 주석 해제
    # "JO": "민법",     # 참조법령명 필터(정확도 올리고 싶으면)
}
```

```python
res = requests.get(BASE_URL, params=params, timeout=30)
res.raise_for_status()  # () 꼭!

data = res.json()
```

```python
print(json.dumps(data, indent=2, ensure_ascii=False))
```
